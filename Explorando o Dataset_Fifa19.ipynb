### Vamos utilizar alguns filtros para localização de informações no dataframe. No exemplo em questão, usaremos o loc e iloc eu seguida. Buscarei explica-los da forma mais claro possível:
### - loc => precisamos informar quais linhas queremos exibir os dados e o nome da coluna em que os dados em questão estão  presentes.
### - iloc  => informando as posições (o intervalo que queremos coletar as informações ) e posição da coluna (o intervalo que queremos coletar as informações).

```python
#Demonstração abaixo:
fifa.loc[0:3,'Name'] #nome do dataframe.método[intervalo selecionado,'Nome da coluna']#
```

```python
#Demonstração abaixo:
fifa.iloc[0:3,0:4] #nome do dataframe.método[intervalo selecionado (linhas/rows),intervalo selecionado(colunas/columns)]#
```

### Uma coisa que podemos usar para implementar esses dois métodos de filtragem de informação é usando condicionais. Segue abaixo uma demosntração:

```python
fifa.loc[(fifa['Age']>30) & (fifa['Overall']>90)] # observe o index das linhas, evidenciado que selecionamos valores 
# com essas condições.
```

### Vamos avançar um pouco no desenvolvimento do nosso notebook. Iremos pegar uma coluna que contenha um dado numérico e um string, na mesma 'célula', usando o termo do Excel para facilitar a compreensão. Vamos substituir o texto (str) por um espaço vazio. Mas antes, vamos substituir todos os dados NaN por um dado numérico e uma str padrão, nesse caso a nossa coluna analisada é a de peso (weight) tem a unidade em libras(lbs) na mesma célula que o peso, faremos essa substituição. Usando o replace em conunto com uma função, sendo aplicada no nosso dataframe.

```python
# Consultando a coluna Weight (Peso) e mostrando sua estrutura.
fifa['Weight'].head(3)
```

```python
# Inserindo nas células NaN (nulas) da coluna Weight, 0 lbs .
fifa['Weight'].fillna('0lbs',inplace=True)
```

```python
# Criamos uma função para tratar o campo peso e 
# pedimos para retornar um valor inteiro, para 
# poderia ser um flutuante também.

def subs_lbs (value):
    
    x = value.replace('lbs',' ')
    return int(x)
fifa['Weight_New'] = fifa['Weight'].apply(subs_lbs)
fifa.head(5)
```
